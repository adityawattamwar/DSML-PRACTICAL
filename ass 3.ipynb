```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "D://DSML PRACTICAL//Datasets//House Data.csv"
df = pd.read_csv(file_path)
num_df= df.select_dtypes(include=['number'])

# Calculate standard deviation for each feature
std_dev = num_df.std()

# Calculate variance for each feature
variance = num_df.var()

# Calculate percentiles (25th, 50th, and 75th) for each feature
percentiles = num_df.quantile([0.25, 0.5, 0.75])

# Display results
print("Standard Deviation for each feature:")
print(std_dev)

print("\nVariance for each feature:")
print(variance)

print("\nPercentiles (25%, 50%, 75%) for each feature:")
print(percentiles)
import seaborn as sns
# Create histograms for each feature
for column in df.columns:
    sns.histplot(df[column])
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
    
```
